```python
df = pd.read_csv("dummy.csv",encoding = 'euc-kr')
df
```

```python
ko = []
ameri = []
for i in range(len(df)):
    if df.iloc[i, 2] == "한국":
        ko += "1"
    else:
        ko += "0"
    if df.iloc[i, 2] == "미국":
        ameri += "1"
    else:
        ameri += "0"
Nation_df = pd.DataFrame({'ko' : ko, 'ameri' : ameri})
Nation_df
```

```python
genre_df = pd.DataFrame({'A' : []})
x = ["액션", "어드벤처", "범죄", "스릴러"]
for x_genre in x:
    col = []
    for i in range(len(df)):
        ch = df.iloc[i, 1].split(",")
        if ch.count(x_genre) == 1:
            col += '1'
        else:
            col += "0"
    col = pd.DataFrame({x_genre : col})
    genre_df = pd.concat([genre_df, col], axis=1)
genre_df
```
