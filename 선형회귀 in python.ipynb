```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
import statsmodels.api as sm
import os
import pandas as pd
```

```python
os.getcwd()
os.chdir('C:\\Users\\user\\Desktop')
```

```python
train = pd.read_csv("movie_data.csv",encoding = 'euc-kr')
test = pd.read_csv("test_data.csv", encoding = 'euc-kr')
train
```

```python
y_train = train.iloc[:,1]
x_train = train.iloc[:,2:]
y_test = test.iloc[:,1]
x_test = test.iloc[:,2:]
```

```python
#sm.OLS(frist_week_audience, first_day_screen+drama+action, data = train)
model = sm.OLS(y_train, x_train)
result = model.fit()
print(result.summary())
#Durbin-Watson : 오차항끼리의 상관관계를 나타낸다. 2에 가까울 수록 상관관계가 없다.
```

```python
new = result.predict(x_test)
new
```

```python
new_x_train = x_train.iloc[:, [0, 2, 8, 16, 23]]
new_x_test = x_test.iloc[:, [0, 2, 8, 16, 23]]
new_x_train
```

```python
model = sm.OLS(y_train, new_x_train)
new_result = model.fit()
new1 = new_result.predict(new_x_test)
new1
```
