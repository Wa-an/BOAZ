## regularization은 다른 패키지에서도 지원하지만
## 유명한 기계 학습 패키지인 scikit-learn 를 이용하자.

```python
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, Lasso, Ridge, ElasticNet
from sklearn.cross_validation import train_test_split
```

```python
os.getcwd()
```

```python
os.chdir('C:\\Users\\user\\Desktop')
```

```python
sonar = pd.read_csv('Sonar.csv')
sonar
```

```python
x = sonar.drop('Class',axis=1)
x
```

```python
y=[]
for i in sonar.ix[:,'Class'] :
    if i=='R':
        y.append(0)
    if i=='M':
        y.append(1)
y
```

```python
#sklearn.linear_model.LogisticRegression은 기본적으로 regularization 옵션을 내장.
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
#데이터를 7:3의 비율로, train과 test로 나눔
```

```python
x_train
```

```python
model = LogisticRegression()
model.fit(x_train, y_train) 
#penalty='l2' = ridge
```

```python
model.score(x_train,  y_train)
```

```python
model.score(x_test, y_test) #valudation으로 과적합인지 아닌지를 확인가능
```

```python
model.coef_ #이미 정규화 되어있어 계수가 크지 않은 것 또한 볼 수 있다.
```

```python
model_lasso = LogisticRegression(penalty='l1')
model_lasso.fit(x_train, y_train)
```

```python
model_lasso.score(x_train,  y_train)
```

```python
model_lasso.score(x_test, y_test) #valudation으로 과적합인지 아닌지를 확인가능
```

```python
model_lasso.coef_ #이미 정규화 되어있어 계수가 크지 않은 것 또한 볼 수 있다.
#결론
#Ridge Regression이 Sonar 데이터에 가장 적합하다
```
